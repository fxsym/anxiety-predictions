```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
le = LabelEncoder()

# Mengubah data kategori menjadi numerik
data['Smoking'] = le.fit_transform(data['Smoking'])
data['Dizziness'] = le.fit_transform(data['Dizziness'])
data['Recent Major Life Event'] = le.fit_transform(data['Recent Major Life Event'])

# Untuk 'Gender', kita bisa melakukan One-Hot Encoding
data = pd.get_dummies(data, columns=['Gender'], drop_first=True)

# Cek data setelah encoding
print(data.head())
```

```python
from sklearn.model_selection import train_test_split

# Pilih fitur numerik yang relevan (pastikan nama kolom sama persis)
X = data[[
    "Caffeine Intake (mg/day)",
    "Stress Level (1-10)",
    "Heart Rate (bpm)",
    "Breathing Rate (breaths/min)",
    "Sweating Level (1-5)",
    "Sleep Hours",
    "Alcohol Consumption (drinks/week)",
    "Smoking",
    "Dizziness",
    "Recent Major Life Event",
    "Gender_Male"  # Jika menggunakan One-Hot Encoding, pastikan nama kolom sesuai
]]
y = data["Anxiety Level (1-10)"]

from sklearn.preprocessing import StandardScaler

# Standarisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data dengan data yang sudah distandarisasi
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_log_error, r2_score
import numpy as np

# Parameter distribusi untuk RandomizedSearchCV
param_dist = {
    'n_estimators': np.arange(100, 1001, 100),
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': np.arange(2, 11),
    'min_samples_leaf': np.arange(1, 11),
    'max_features': ['sqrt', 'log2', None],
    'bootstrap': [True, False]
}

# Inisialisasi model dasar
rf = RandomForestRegressor(random_state=42)

# RandomizedSearchCV untuk tuning hyperparameter
random_search = RandomizedSearchCV(
    rf,
    param_distributions=param_dist,
    n_iter=100,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    random_state=42,
    verbose=1
)

# Training model dengan data yang telah dibersihkan dan distandarisasi
random_search.fit(X_train, y_train)

# Mengambil model terbaik
best_rf_random = random_search.best_estimator_

# Melakukan prediksi dengan model terbaik
y_pred_random = best_rf_random.predict(X_test)

# Evaluasi model
mae_random = mean_absolute_error(y_test, y_pred_random)
rmse_random = mean_squared_log_error(y_test, y_pred_random) ** 0.5
r2_random = r2_score(y_test, y_pred_random)

print(f"MAE (RandomizedSearch): {mae_random:.4f}")
print(f"RMSE (RandomizedSearch): {rmse_random:.4f}")
print(f"R Score (RandomizedSearch): {r2_random:.4f}")
```

```python
import matplotlib.pyplot as plt

# Visualisasi Prediksi vs Aktual
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_random, color='dodgerblue', edgecolor='k', alpha=0.7, label='Predictions')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2, linestyle='--', label='Ideal Line')
plt.xlabel("Actual Anxiety Level", fontsize=12)
plt.ylabel("Predicted Anxiety Level", fontsize=12)
plt.title("Actual vs Predicted Anxiety Level (Random Forest)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
